```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pyplot
import scipy.stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
from js import fetch
import io


URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
resp = await fetch(URL)
boston_url = io.BytesIO((await resp.arrayBuffer()).to_py())
```

```python
boston_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
boston_df=pd.read_csv(boston_url)
```

### Identifying Duplicates

```python
print('Number of duplicate rows:', boston_df.duplicated().sum())
if boston_df.duplicated().sum() > 0:
    print('\nDuplicate Rows:')
    display(boston_df[boston_df.duplicated()])
else:
    print('No duplicate rows found.')
```

### Plotting Scatterplots

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x='RM', y='MEDV', data=boston_df)
plt.title('Scatterplot of RM vs MEDV')
plt.xlabel('Average number of rooms (RM)')
plt.ylabel('Median value of homes (MEDV)')

plt.subplot(1, 2, 2)
sns.scatterplot(x='LSTAT', y='MEDV', data=boston_df)
plt.title('Scatterplot of LSTAT vs MEDV')
plt.xlabel('% lower status of the population (LSTAT)')
plt.ylabel('Median value of homes (MEDV)')

plt.tight_layout()
plt.show()
```

### Plotting Boxplots for Categorical Features

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='CHAS', y='MEDV', data=boston_df)
plt.title('Boxplot of MEDV by CHAS (Charles River dummy variable)')
plt.xlabel('CHAS (1 if tract borders river; 0 otherwise)')
plt.ylabel('Median value of homes (MEDV)')

plt.subplot(1, 2, 2)
sns.boxplot(x='RAD', y='MEDV', data=boston_df)
plt.title('Boxplot of MEDV by RAD (index of accessibility to radial highways)')
plt.xlabel('RAD (index of accessibility to radial highways)')
plt.ylabel('Median value of homes (MEDV)')

plt.tight_layout()
plt.show()
```

### Data Visualization: Distributions of Key Variables

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot for MEDV
plt.figure(figsize=(8, 6))
sns.boxplot(y=boston_df['MEDV'])
plt.title('Boxplot of MEDV (Median value of owner-occupied homes)')
plt.ylabel('MEDV ($1000s)')
plt.show()
```

```python
# Histograms for CRIM, MEDV, and RM
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
sns.histplot(boston_df['CRIM'], kde=True)
plt.title('Histogram of CRIM')

plt.subplot(1, 3, 2)
sns.histplot(boston_df['MEDV'], kde=True)
plt.title('Histogram of MEDV')

plt.subplot(1, 3, 3)
sns.histplot(boston_df['RM'], kde=True)
plt.title('Histogram of RM')

plt.tight_layout()
plt.show()
```
